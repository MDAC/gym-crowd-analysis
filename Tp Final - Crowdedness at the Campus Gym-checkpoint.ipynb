## Estimar y hacer predicciones

#### 17. Sacamos unas columnas:

```python
data_copy = data.drop(['date','timestamp'],axis=1)
```

```python
data_copy.head()
```

#### 18. Definimos parámetros X, y:

```python
y = data_copy['number_people'].values # nuestro target
data_copy = data_copy.drop(['number_people'],axis=1)
X = data_copy.values # las features
data_copy.head()
```

#### 19. Dividimos el dataset en training set y testing set:

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=20)
```

#### 20. Estandarizamos y escalamos el dataset entre -1 y 1:

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)  
X_test = scaler.transform(X_test) 
```

#### 21. Analizamos la importancia de las features usando RandomForestRegressor:

```python
from sklearn.ensemble import RandomForestRegressor
import numpy as np # linear algebra
```

```python
rdm = RandomForestRegressor()
%time rdm.fit(X_train, y_train)
y_pred = rdm.predict(X_test)
print(rdm.score(X_test, y_test))
```

```python
indices = np.argsort(rdm.feature_importances_)[::-1]

print(u'Ranking de los variables según su importancia:')

for f in range(data_copy.shape[1]):
    print('%d. Variable %d %s (%f)' % (f+1 , indices[f], data_copy.columns[indices[f]],
                                      rdm.feature_importances_[indices[f]]))
```

#### 22. Analizamos cómo varía score según distintos n_estimators de RandomForestRegressor:

```python
rdm = RandomForestRegressor(n_jobs=-1) #define el modelo

# Probando con diferentes n_estimators - esto tomará un minuto o más
estimators = np.arange(10, 200, 10) #entre rango 10 a 200, con intervalo de 10
scores = []
for n in estimators:
    rdm.set_params(n_estimators=n)
    rdm.fit(X_train, y_train)
    scores.append(rdm.score(X_test, y_test))
```

```python
plt.figure(figsize=(12,8))
plt.xlabel("n_estimator")
plt.ylabel("score")
plt.plot(estimators, scores)
```

#### 23. Apliquemos PCA, para reducir dimensionalidad:

```python
from sklearn.decomposition import PCA

pca = PCA()  #define un objeto PCA
X_train = pca.fit_transform(X_train)  
X_test = pca.transform(X_test)
```

```python
explained_variance = pca.explained_variance_ratio_ #varianzas de cada componentes
explained_variance
```

```python
plt.figure(figsize=(10,8))
ax = sb.barplot(range(8), explained_variance)
ax.set_ylabel('Explained variance')
ax.set_xlabel('Principales componentes')
```

Resulta  los componentes de índices 0 a 6 son responsable de una variación de 96%, y el componente 7 contiene una variación menor que 5%, por lo tanto podemos eliminarlos.

```python
pca = PCA(n_components=7) #reduce la dimensionalidad
X_train = pca.fit_transform(X_train)
X_test = pca.transform(X_test)
```

```python
rdm = RandomForestRegressor()
%time rdm.fit(X_train, y_train) #vuelve a predecir
y_pred = rdm.predict(X_test)
print(rdm.score(X_test, y_test))
```
